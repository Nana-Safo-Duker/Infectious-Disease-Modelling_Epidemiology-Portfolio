# Detailed Notebook: Deep learning vector diseases


## Setup


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from pathlib import Path
np.random.seed(303)
```

## Load or Create Dataset


```python
data_path = Path('data/dataset.csv')
if data_path.exists():
    df = pd.read_csv(data_path)
else:
    n = 66
    half = n // 2
    group_a = np.random.normal(48, 9, half)
    group_b = np.random.normal(48 + 1, 9, n - half)
    df = pd.DataFrame({
        'time': np.arange(1, n + 1),
        'signal': np.concatenate([group_a, group_b]),
        'group': np.array(['A'] * half + ['B'] * (n - half))
    })
df.head()
```

## Summary Statistics and T-test


```python
stats_df = pd.DataFrame({
    'mean': [df['signal'].mean()],
    'median': [df['signal'].median()],
    'std': [df['signal'].std()]
})
a = df.loc[df['group'] == 'A', 'signal']
b = df.loc[df['group'] == 'B', 'signal']
t_stat, p_val = stats.ttest_ind(a, b, equal_var=False)
stats_df, {'t_statistic': float(t_stat), 'p_value': float(p_val)}
```

## Plot


```python
plt.figure(figsize=(10, 4))
for g, sub in df.groupby('group'):
    plt.plot(sub['time'], sub['signal'], label=f'Group {g}', alpha=0.8)
plt.title('Signal Over Time by Group')
plt.xlabel('Time')
plt.ylabel('Signal')
plt.legend()
plt.tight_layout()
plt.show()
```
